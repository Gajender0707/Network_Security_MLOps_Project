```python
train_data=pd.read_csv("/Users/sanju/Documents/DS/MLOps/MLOps_Projects/Network_Security_Project/artifacts/data_ingestion/ingested/train.csv")
test_data=pd.read_csv("/Users/sanju/Documents/DS/MLOps/MLOps_Projects/Network_Security_Project/artifacts/data_ingestion/ingested/test.csv")
```

```python
from scipy.stats import ks_2samp
```

```python
ks_2samp(train_data,test_data,alternative='two-sided',method="auto")
```

```python
import numpy as np
```

```python
report={}
status=None
for column in train_data.columns:
    d1=train_data[column]
    d2=test_data[column]
    # print(f"this si d1: {d1}")
    # print(f"This is d2: {d2}")
    is_same_dist=ks_2samp(d1,d2)
    if is_same_dist.pvalue>0.05:
        status=True
    else:
        status=False
    report.update({column:{
        "report_status":status,
        "p_value":float(is_same_dist.pvalue)
    }})
```
